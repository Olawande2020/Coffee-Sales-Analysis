# Done with data cleaning now using data to provide insights

#### SALES TRENDS

```python
#How do coffee sales change over time for both stores?
store_1["month"] = store_1["date"].dt.to_period("M")
monthly_sales_1 = store_1.groupby("month")["money"].sum()
monthly_sales_1.sort_values(ascending=False)
```

```python
store_2["month"] = store_2["date"].dt.to_period("M")
monthly_sales_2 = store_2.groupby("month")["money"].sum()
monthly_sales_2.sort_values(ascending=False)
```

```python
#Which days have the highest sales for both stores?
daily_sales = store_1.groupby("date")["money"].sum()
daily_sales.sort_values(ascending=False).head()
```

```python
daily_sales = store_2.groupby("date")["money"].sum()
daily_sales.sort_values(ascending=False).head()
```

```python
# What time of day do people buy coffee most in both stores?
store_1["datetime"] = pd.to_datetime(store_1["datetime"])
store_1["hour"] = store_1["datetime"].dt.hour

store_1.groupby("hour")["money"].sum().sort_values(ascending=False)
```

```python
store_2["datetime"] = pd.to_datetime(store_2["datetime"])
store_2["hour"] = store_2["datetime"].dt.hour

store_2.groupby("hour")["money"].sum().sort_values(ascending=False)
```
